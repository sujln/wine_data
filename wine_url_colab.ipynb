```python
headers = {'User-Agent': 'Mozilla/5.0'} 
```

```python
review_all=[]
wine_id=[]
```

```python
count = 1

for url in white_url[5000:]:
  try:
    if 'year=' in str(url):
      review_url = 'https://www.vivino.com/api/wines/' + url.split('w/')[1].split('?')[0] + '/reviews?per_page=10&page=2&year=' + url.split('year=')[1].split('&price')[0]
    else:
      review_url = 'https://www.vivino.com/api/wines/' + url.split('w/')[1].split('?')[0] + '/reviews?per_page=10&page=2&year=N.V'
    web = requests.get(review_url, headers=headers).content
    review = web.decode('utf8')
    my_review = json.loads(review)
  except:
    print(url)
    continue

  wine_id.append(url.split('id=')[1])
  page_review = []
  for i in range(len(my_review['reviews'])):
    page_review.append(my_review ['reviews'][i]['note'])
  review_all.append(page_review)

  
  print(count)
  count+=1
```

```python
white_review = pd.DataFrame({'id' : wine_id,
                  're' :review_all})
```

# 와인 정보 뽑기

```python
red_mid_url = pd.read_excel('red_wine_mid_url.xlsx')
```

```python
red_mid_url = red_mid_url.iloc[:,1]
```

```python
red_mid_url =red_mid_url.to_list()
```

```python
len(red_mid_url)
```

```python
headers = {'User-Agent': 'Mozilla/5.0'} 
```

```python
num=0
```

```python
name = []
price =[]
score=[]
winery = []
grapes=[]
country=[]
region=[]
#wine_style=[]  
alcohol=[]
# allergens=[]
# light_bold=[]
# smooth_tannic=[]
# dry_sweet=[]
# soft_acidic=[]
# notes=[]
foods=[]
wine_id=[]
```

```python
count = 1

for url in red_mid_url[1000:2000]: #너무 많아서 1000개씩 쪼개주면서 함
      web = requests.get(url, headers=headers).content
      source = BeautifulSoup(web, 'html.parser')
      try:
        for idx, script in enumerate(source.find_all('script')):
          if 'alcohol' in str(script):
            num=idx
        data = source.find_all('script')[num]
        data = data.get_text()
        data = data.split('vintagePageInformation = ')[1].split(';\n  window.__PRELOADED_STATE__')[0]
        data = json.loads(data)
  
      except:
        continue

      wine_id.append(url.split('id=')[1])
      name.append(source.find('span',{'class':'vintage'}).get_text().replace("\n"," ").strip())

      try:
        if 'price' in data:
            price.append(data['price']['amount'])
        else:
            price.append(data['prices_and_availability']['availability']['median']['amount'])
      except:
        price.append("")



      score.append(data['vintage']['statistics']['ratings_average'])
      winery.append(data['vintage']['wine']['winery']['name'])

      if data['vintage']['wine']['grapes']!=[]:
          grapes.append(data['vintage']['wine']['grapes'][0]['name'])
      else:
          grapes.append("")
      try:
        country.append(data['vintage']['wine']['region']['country']['name']) 
        region.append(data['vintage']['wine']['region']['name'])
      except:
        country.append("")
        region.append("")
      
      try:
      #if ('alcohol' in str(data['vintage']['wine_facts'])) & (data['vintage']['wine_facts'] != {}):        
          alcohol.append(data['vintage']['wine_facts']['alcohol'])
      except:
      #else:
          alcohol.append("")

      food = []    
      if 'foods' in str(data['vintage']['wine']):
          for i in range(len(data['vintage']['wine']['foods'])):
              food.append(data['vintage']['wine']['foods'][i]['name'])
          foods.append(food)
      else:
          foods.append("")

    
      print(count)
      count+=1
```

```python
len(wine_id)
```

```python
wine_df= pd.DataFrame({'name':name,
                       'price': price,
                       'score':score,
                       'winery':winery,
                       'grapes':grapes,
                       'country':country,
                       'region':region,
                       'alcohol':alcohol,
                       'foods':foods,
                       'wine_id':wine_id
                       })
```

```python
wine_df
```

```python
wine_df.to_csv("red_wine_mid_info_0_2000.csv")
```
